````python
import pandas as pd
import numpy as np

# Make results reproducible
np.random.seed(42)

genes = [
    "TP53","BRCA1","EGFR","KRAS","MYC","ALK","BRAF","PIK3CA","CDK4","PTEN",
    "RB1","CCND1","ERBB2","MET","FGFR1","VEGFA","CDH1","SMAD4","APC","NRAS",
    "AKT1","MTOR","NOTCH1","JAK2","STAT3","NFKB1","HIF1A","MDM2","TERT","CDKN2A",
    "ATM","ATR","CHEK1","CHEK2","BCL2","BAX","CASP3","FAS","MMP9","CXCL8",
    "IL6","TGFB1","PDCD1","CD274","CTLA4","LAG3","TIGIT","IFNG","GZMB","PRF1",
    "SOX2","NANOG","POU5F1","VIM","EPCAM","SNAI1","TWIST1","ZEB1","MUC1","KRT18",
    "KRT19","COL1A1","FN1","ITGA6","ITGB1","CXCR4","CCR7","MKI67","PCNA","TOP2A",
    "GATA3","FOXA1","ESR1","AR","RUNX1","IDH1","IDH2","FLT3","KIT","ROS1",
    "RET","NTRK1","NTRK2","NTRK3","SMARCA4","ARID1A","KEAP1","STK11","NF1","NF2",
    "BRIP1","PALB2","RAD51","XRCC1","MSH2","MSH6","MLH1","PMS2","POLE","POLD1"
]

n = len(genes)

df = pd.DataFrame({
    "Gene": genes,
    "Expression": np.round(np.random.uniform(5, 25, n), 2),
    "Cancer_Type": ["LUAD"] * n,
    "Chromosome": np.random.randint(1, 23, n),
    "Gene_Type": np.random.choice(
        ["Oncogene", "Tumor Suppressor"],
        n
    ),
    "Log2FC": np.round(np.random.uniform(-4, 4, n), 2),
    "P_Value": np.round(np.random.uniform(0.0001, 0.05, n), 4),
    "Adjusted_P_Value": np.round(np.random.uniform(0.0001, 0.05, n), 4)
})

df["Status"] = np.where(df["Log2FC"] > 0,
                        "Upregulated",
                        "Downregulated")

df.head()
````

````python
# Display descriptive statistics

df.describe()
````

# **Understanding the Output**

The `describe()` function automatically calculates summary statistics for all numerical columns in the dataset.

For the **Expression**, **Chromosome**, **Log2FC**, **P_Value**, and **Adjusted_P_Value** columns, Pandas calculates:

- Count
- Mean
- Standard Deviation
- Minimum
- 25th Percentile
- Median (50%)
- 75th Percentile
- Maximum

These values provide an overall summary of the dataset.

# **Explain Every Statistic**

**Count**


Count represents the total number of observations available in a numerical column.

For example,

**Count** = 100

means that expression values are available for all 100 genes.

If the count is smaller than expected, it usually indicates missing values.

# **Mean**

The mean is the average value.

**Formula:**

Mean = Sum of all observations ÷ Number of observations

In gene expression studies, the mean represents the average expression level across all genes.

Researchers often compare mean expression levels between healthy and diseased tissues.

# **Standard Deviation (Std)**

Standard deviation measures variability.

A small standard deviation means most genes have similar expression levels.

A large standard deviation indicates substantial variation between genes.

High variability often suggests biologically interesting genes that may require further investigation.

# **Minimum**

The minimum value represents the lowest observed expression level.

Genes with extremely low expression may be:

- weakly expressed
- inactive
- tissue specific

# **Maximum**

The maximum value represents the highest observed expression level.

Highly expressed genes often become candidates for:

- biomarkers
- therapeutic targets
- pathway analysis

# **25th Percentile (First Quartile)**

Twenty-five percent of observations fall below this value.

This statistic helps researchers understand the lower range of gene expression.

# **50th Percentile (Median)**

The median is the middle value.

Half of all observations are below the median.

Unlike the mean, the median is not strongly affected by extreme values.

For skewed biological datasets, the median often provides a better estimate of the central tendency.

# **75th Percentile (Third Quartile)**

Seventy-five percent of observations fall below this value.

Genes above this percentile are generally considered relatively highly expressed compared with the rest of the dataset.

````python
#calculate each statistic individually.

print("Mean Expression:")
print(df["Expression"].mean())

print("\nMedian Expression:")
print(df["Expression"].median())

print("\nStandard Deviation:")
print(df["Expression"].std())

print("\nMinimum Expression:")
print(df["Expression"].min())

print("\nMaximum Expression:")
print(df["Expression"].max())
````

# **Output Explanation**

Instead of relying only on `describe()`, Pandas allows researchers to calculate each statistic individually.

This approach is particularly useful when generating automated reports or comparing multiple biological datasets.

````python
#Calculate the quartiles separately.

print("25th Percentile:")
print(df["Expression"].quantile(0.25))

print("\nMedian:")
print(df["Expression"].quantile(0.50))

print("\n75th Percentile:")
print(df["Expression"].quantile(0.75))
````

# **Biological Interpretation**

Suppose the output shows:

Mean Expression = 15.8

This indicates that the average normalized expression level across the analyzed LUAD-associated genes is approximately 15.8.

Genes with expression values substantially greater than the mean may represent highly active genes and could be investigated further as potential biomarkers or therapeutic targets.

Similarly, genes with very low expression may represent inactive or tissue-specific genes.

The `dropna()` function removes every row that contains one or more missing values.

This creates a cleaned version of the dataset that can be used for downstream analyses.

However, removing too many rows may reduce statistical power if the dataset contains many missing observations.

````python
# Fill missing Expression values using the mean

df["Expression"] = df["Expression"].fillna(
    df["Expression"].mean()
)

df.head()
````

# **Why Use the Mean?**

Replacing missing values with the mean preserves the size of the dataset while maintaining the overall average expression level.

This approach is commonly used when only a few values are missing and the data are approximately normally distributed.

However, excessive imputation should be avoided because it may reduce biological variability.

````python
# Fill missing Log2FC values using the median

df["Log2FC"] = df["Log2FC"].fillna(
    df["Log2FC"].median()
)
````

# **Why Use the Median?**

The median is less affected by extreme values than the mean.

For skewed biological datasets, the median often provides a more robust estimate when replacing missing observations.

````python
# Replace missing adjusted p-values

df["Adjusted_P_Value"] = df["Adjusted_P_Value"].fillna(0.05)
````

# **Biological Interpretation**

Adjusted p-values estimate the statistical significance of differential expression after correcting for multiple hypothesis testing.

Assigning a conservative value such as **0.05** prevents missing observations from being incorrectly interpreted as statistically significant.

In real research projects, the method used for replacing missing p-values depends on the experimental design and statistical analysis pipeline.

## **Explanation**

 **Why Adjusted P-values?**

Thousands of genes are tested simultaneously during RNA-Seq analysis.

Multiple statistical tests increase the probability of false positive results.

Adjusted p-values correct for this problem.

Genes with Adjusted P-value < 0.05 are generally considered statistically significant.

These genes become candidates for downstream biological interpretation.

````python
#Count Significant Genes
print("Number of Significant Genes:")

print(len(significant_genes))
````

````python
#Candidate Biomarkers
candidate_biomarkers = df[
    (df["Expression"]>18) &
    (df["Adjusted_P_Value"]<0.05)
]

candidate_biomarkers
````

# **Best Practices for Biomedical Data Analysis**

Professional computational biologists generally follow these best practices:

**Always inspect the dataset before analysis.**

Use:

- head()
- tail()
- shape
- info()

to understand the dataset.



**Never ignore missing values.**

Investigate why data are missing before deciding whether to remove or replace observations.



**Use meaningful variable names.**

Avoid generic names such as:

data1

table

Instead use names such as:

gene_expression

clinical_data

candidate_biomarkers



**Keep raw data unchanged.**

Always create cleaned copies rather than modifying the original dataset directly.



**Save intermediate results.**

Export filtered datasets as CSV files to improve reproducibility.



**Document every analysis.**

Explain what each code cell accomplishes and why it is biologically meaningful.

Common Mistakes Made by Beginners

When starting computational biology with Pandas, students often make mistakes such as:

* Ignoring missing values.

* Using incorrect column names.

* Forgetting to verify data types.

* Filtering genes without considering statistical significance.

* Modifying the original dataset instead of creating a cleaned copy.

* Performing analyses without documenting biological interpretations.

Avoiding these mistakes improves both computational accuracy and scientific reproducibility.
