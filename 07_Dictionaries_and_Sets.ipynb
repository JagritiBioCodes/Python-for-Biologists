## Sets in Python

A **set** is an unordered collection of unique elements.

Unlike lists, sets automatically remove duplicate values. This property is particularly useful in computational biology, where we often need to identify unique genes, proteins, DNA sequences, or mutations from large biological datasets.

For example, if the same gene appears multiple times in sequencing results, converting the data into a set helps retain only the unique gene names.

```python
#SECTION 2
#Sets
#Example
genes = {"TP53","BRCA1","EGFR","TP53","MYC"}

print(genes)
```

### Explanation

Although **TP53** was entered twice, it appears only once in the output because sets do not allow duplicate elements.

Output:

{'TP53', 'BRCA1', 'EGFR', 'MYC'}

This demonstrates that Python automatically removes duplicate values when creating a set.

### Biological Significance

During RNA-Seq or DNA sequencing experiments, the same gene may be detected multiple times.

Researchers often need a list of **unique genes** before performing downstream analyses such as:

- Differential Gene Expression Analysis
- Gene Ontology (GO) Enrichment
- Pathway Analysis
- Biomarker Identification

Python sets provide an efficient way to remove duplicate gene names from biological datasets.

```python
# some more examples
sequencing_results = [
    "TP53",
    "BRCA1",
    "EGFR",
    "TP53",
    "MYC",
    "EGFR",
    "KRAS"
]

print("Original Results:")
print(sequencing_results)

unique_genes = set(sequencing_results)

print("\nUnique Genes:")
print(unique_genes)
```

# **Explanation**
The sequencing experiment reported TP53 and EGFR multiple times.

After converting the list into a Set, duplicate observations were removed automatically.

This leaves only the unique genes that can be used for downstream biological analyses.

# **Research Note**
Modern RNA-Seq experiments may identify more than 20,000 expressed genes.

Before pathway enrichment or differential gene expression analysis, duplicate gene identifiers are removed.

Python Sets provide one of the fastest methods for obtaining unique biological observations.

# **What is Intersection?**
Intersection is a Set operation that identifies elements common to two or more Sets.

Mathematically,

A ∩ B

represents all elements present in both Set A and Set B.

In Computational Biology, Intersection is widely used to identify:

• Shared genes

• Common mutations

• Overlapping biomarkers

• Conserved proteins

• Common biological pathways

```python
#Example 2 Basic Intersection
sample1 = {"TP53", "BRCA1", "EGFR"}

sample2 = {"EGFR", "KRAS", "MYC"}

common_genes = sample1.intersection(sample2)

print(common_genes)
```

# **Explanation**
Python compared the two Sets.

Only EGFR was present in both samples.

Therefore, the Intersection contains only EGFR.

# **Biological Interpretation**
Suppose two lung cancer patients undergo mutation analysis.

Patient 1 mutations:

TP53
BRCA1
EGFR

Patient 2 mutations:

EGFR
KRAS
MYC

EGFR is present in both patients.

This may indicate:

• A shared oncogenic driver mutation

• A common therapeutic target

• A potential biomarker for further investigation

Researchers often perform this type of comparison before selecting targeted therapies or conducting biomarker validation studies.

```python
#Biological Example – Comparing Two Cancer Patients
patient_A = {

    "TP53",

    "EGFR",

    "ALK",

    "KRAS"

}

patient_B = {

    "EGFR",

    "BRAF",

    "KRAS",

    "ROS1"

}

common_mutations = patient_A.intersection(patient_B)

print("Common Mutations:")

print(common_mutations)
```
