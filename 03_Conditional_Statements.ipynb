# **2. If-else Statement**


```python
dna_length = 850

if dna_length >= 1000:
    print("Long DNA sequence")
else:
    print("Short DNA sequence")
```

# **3. If-elif-else Statement**

```python
gc_content = 58

if gc_content > 60:
    print("High GC Content")
elif gc_content >= 40:
    print("Moderate GC Content")
else:
    print("Low GC Content")
```

# **4. Comparison Operators**

```python
geneA = 150
geneB = 210

print(geneA == geneB)
print(geneA != geneB)
print(geneA > geneB)
print(geneA < geneB)
print(geneA >= geneB)
print(geneA <= geneB)
```

# **5. Logical Operators**

```python
gc_content = 55
length = 1800

if gc_content > 50 and length > 1000:
    print("Sequence suitable for downstream analysis")
```

```python
gene_detected = True
protein_detected = False

if gene_detected or protein_detected:
    print("Biological signal detected")
```

```python
contamination = False

if not contamination:
    print("Sample passed quality control")
```

# **6. Nested if Statement**

```python
gc_content = 61
length = 2500

if gc_content > 50:
    if length > 2000:
        print("High-quality genomic sequence")
```

# **Mini Biological Example**

```python
sequence = "ATGCGCGCGCGATAT"

gc = (sequence.count("G") + sequence.count("C")) / len(sequence) * 100

print("GC Content:", gc)

if gc > 60:
    print("GC-rich sequence")
else:
    print("AT-rich sequence")
```
