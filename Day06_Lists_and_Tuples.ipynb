## Topic: Lists and Tuples

```python
#List
genes = ["TP53","BRCA1","KRAS","MYC"]

print(genes)
```

```python
#Access
print(genes[0])
print(genes[2])
```

```python
#Append
genes.append("EGFR")

print(genes)
```

```python
#Remove
genes.remove("MYC")

print(genes)
```

```python
#Sort
genes.sort()

print(genes)
```

```python
#Tuple
chromosomes = (1,2,3,4,5)

print(chromosomes)
```

```python
#Loop
for gene in genes:
    print(gene)
```

```python
#Mini Project
gene_expression = [10.5,15.2,18.4,9.6,21.8]

print("Maximum =", max(gene_expression))
print("Minimum =", min(gene_expression))
print("Average =", sum(gene_expression)/len(gene_expression))
```
