# Python Data Types

Python stores different kinds of information using different data types.

The four most commonly used data types are:

- **String (str):** Stores text enclosed in quotation marks.
- **Integer (int):** Stores whole numbers without decimal points.
- **Float (float):** Stores decimal numbers.
- **Boolean (bool):** Stores logical values: True or False.

```python

# Python Data Types in Biology
#EXAMPLE-1


# String
gene_name = "TP53"

# Integer
chromosome_number = 17

# Float
gene_expression = 12.56

# Boolean
tumor_suppressor = True

print("Gene Name:", gene_name)
print("Chromosome:", chromosome_number)
print("Gene Expression:", gene_expression)
print("Tumor Suppressor:", tumor_suppressor)

print()

print(type(gene_name))
print(type(chromosome_number))
print(type(gene_expression))
print(type(tumor_suppressor))
```

```python
#EXAMPLE-2
#Integer (Whole Number)
sample_count = 150

# Float (Decimal Number)
gene_expression = 8.95

# Boolean (True or False)
mutation_detected = True

# Display Values
print("Number of Samples:", sample_count)
print("Gene Expression Value:", gene_expression)
print("Mutation Detected:", mutation_detected)

print()

# Display Data Types
print("Data Type of sample_count:", type(sample_count))
print("Data Type of gene_expression:", type(gene_expression))
print("Data Type of mutation_detected:", type(mutation_detected))
```

```python
#EXAMPLE-3 CANCER GENOMICS
#Integer (Whole Number)
patient_id = 101

# Float (Decimal Number)
tumor_size = 4.75

# Boolean (True or False)
cancer_detected = True

print("Patient ID:", patient_id)
print("Tumor Size (cm):", tumor_size)
print("Cancer Detected:", cancer_detected)

print()

print(type(patient_id))
print(type(tumor_size))
print(type(cancer_detected))
```

```python
#Count Nucleotides
dna = "ATGCGTACTTAGCGAT"

print("A =", dna.count("A"))
print("T =", dna.count("T"))
print("G =", dna.count("G"))
print("C =", dna.count("C"))
```

```python
#Calculate GC Content
dna = "ATGCGTACTTAGCGAT"

g = dna.count("G")
c = dna.count("C")

gc_content = ((g + c) / len(dna)) * 100

print("GC Content =", round(gc_content, 2), "%")
```

```python
#Output cleaner
dna = "ATGCGATACGCTTGA"

gc = ((dna.count("G") + dna.count("C")) / len(dna)) * 100

print(f"GC Content = {gc:.2f}%")
```
