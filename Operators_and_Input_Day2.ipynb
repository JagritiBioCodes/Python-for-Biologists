## Arithmetic Operators

Arithmetic operators perform mathematical calculations.

```python
gene_A = 120
gene_B = 80

print("Addition:", gene_A + gene_B)
print("Subtraction:", gene_A - gene_B)
print("Multiplication:", gene_A * gene_B)
print("Division:", gene_A / gene_B)
print("Modulus:", gene_A % gene_B)
print("Power:", gene_A ** 2)
```

## Comparison Operators

Comparison operators compare two values and return True or False.

```python
sample1 = 250
sample2 = 180

print(sample1 > sample2)
print(sample1 < sample2)
print(sample1 == sample2)
print(sample1 != sample2)
print(sample1 >= sample2)
print(sample1 <= sample2)
```

## Assignment Operators

```python
reads = 1000

reads += 500
print(reads)

reads -= 200
print(reads)

reads *= 2
print(reads)

reads /= 5
print(reads)
```

## Checking Data Types using type()

```python
gene_name = "BRCA1"
gene_length = 1863
gc_content = 53.4
is_cancer_gene = True

print(type(gene_name))
print(type(gene_length))
print(type(gc_content))
print(type(is_cancer_gene))
```

## Type Conversion

```python
reads = input("Enter Number of Reads: ")

print(type(reads))

reads = int(reads)

print(type(reads))

print(reads + 100)
```
