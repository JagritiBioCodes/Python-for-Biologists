# **What is NumPy?**
NumPy (Numerical Python) is a scientific computing library that provides efficient multidimensional arrays and mathematical operations.

Compared with Python lists, NumPy arrays:

• Require less memory

• Perform calculations much faster

• Support vectorized mathematical operations

• Are ideal for handling biological datasets containing thousands of genes.

NumPy serves as the computational engine behind many bioinformatics and machine learning workflows.

```python
#Import NumPy
import numpy as np
```

# **Explanation**
The standard alias **np** is universally used by researchers and data scientists when importing NumPy.

```python
#Example 1 Creating a Gene Expression Array
gene_expression = np.array([

12.4,

8.6,

20.2,

17.8,

15.4,

10.9,

18.6,

9.7

])

print(gene_expression)
```

# **Explanation**
Each value represents the expression level of a specific gene.

Unlike Python lists, NumPy arrays are optimized for fast numerical computation and mathematical analysis.

# **Biological Interpretation**
Suppose these values represent normalized expression levels of eight cancer-related genes measured using RNA sequencing.

Researchers routinely analyze such numerical arrays before identifying differentially expressed genes or potential biomarkers.

```python
#Example 2 Mean Gene Expression
print(np.mean(gene_expression))
```

# **Explanation**
The mean provides the average expression level across all genes.

Average expression is frequently reported during quality control and exploratory data analysis.

```python
#Example 3 – Median Expression
print(np.median(gene_expression))
```

# **Biological Interpretation**
The median is less affected by extreme values and often provides a more robust measure of central tendency in biological datasets.

```python
#Maximum Expression
print(np.max(gene_expression))
```

```python
#Minimum Expression
print(np.min(gene_expression))
```

```python
#Standard Deviation
print(np.std(gene_expression))
```

# **Explanation**
Standard deviation measures how much expression levels vary among genes.

Higher variation often indicates biological diversity or technical variability.

```python
#Array Arithmetic
normalized = gene_expression / 10

print(normalized)
```

# **Biological Interpretation**
Normalization rescales expression values.

Normalization is a crucial preprocessing step before clustering, machine learning, and differential expression analysis.

```python
#Random RNA-Seq Dataset
reads = np.random.randint(100,1000,20)

print(reads)
```

# **Explanation**
This simulates sequencing read counts obtained from a hypothetical RNA sequencing experiment.

Although randomly generated, it resembles the structure of real count data.

```python
#Biological Example
genes = np.array([

"TP53",

"BRCA1",

"EGFR",

"KRAS",

"MYC",

"ALK",

"BRAF",

"PIK3CA"

])

expression = np.array([

12.4,

8.6,

20.2,

17.8,

15.4,

10.9,

18.6,

9.7

])

for gene, exp in zip(genes, expression):

    print(gene, ":", exp)
```

# **Biological Interpretation**
This example links each gene to its corresponding expression level.

This structure closely resembles simplified gene expression matrices used in transcriptomics studies.

# **Mini Exercise**
A researcher measured expression levels for six immune checkpoint genes.

Expression values:

7.5

12.8

18.9

22.1

15.3

9.6

Tasks:

1. Create a NumPy array.

2. Calculate:

• Mean

• Median

• Maximum

• Minimum

• Standard deviation

3. Normalize the values by dividing each value by 10.

```python
#Solution
immune_expression = np.array([

7.5,

12.8,

18.9,

22.1,

15.3,

9.6

])

print("Mean:", np.mean(immune_expression))

print("Median:", np.median(immune_expression))

print("Maximum:", np.max(immune_expression))

print("Minimum:", np.min(immune_expression))

print("Standard Deviation:", np.std(immune_expression))

print("Normalized Values:")

print(immune_expression/10)
```
