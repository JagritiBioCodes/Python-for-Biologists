# **The Central Dogma of Molecular Biology**

The flow of genetic information within a cell is commonly described by the Central Dogma of Molecular Biology.

```text
        DNA
         │
         ▼
Transcription
         │
         ▼
        RNA
         │
         ▼
Translation
         │
         ▼
      Protein
```

DNA stores genetic information.

During **transcription**, DNA is copied into messenger RNA (mRNA).

During **translation**, the mRNA sequence is decoded to produce proteins, which perform most biological functions within the cell.

Computational analysis of DNA sequences helps researchers understand each step of this process.

# **Part 12-D GC Content Analysis in Computational Biology**
After determining the nucleotide composition of a DNA sequence, one of the next and most informative analyses is the calculation of **GC content**.

GC content represents the percentage of nucleotides within a DNA sequence that are either **Guanine (G)** or **Cytosine (C)**.

Because G and C are connected by **three hydrogen bonds**, whereas A and T are connected by only **two hydrogen bonds**, GC-rich DNA regions are generally more stable.

GC content is therefore an important molecular characteristic that influences DNA structure, stability, PCR amplification, sequencing performance, and genome organization.

In this section, we will calculate GC content manually and using Python, compare multiple DNA sequences, visualize GC percentages, and interpret their biological significance.

# **Why GC Content Matters?**

GC content is routinely calculated in computational biology because it provides insight into several important biological properties.

Researchers use GC content to:

- Evaluate DNA stability.
- Design PCR primers.
- Compare genomes across organisms.
- Characterize genes and chromosomes.
- Assess sequencing quality.
- Investigate evolutionary adaptation.
- Analyze bacterial and viral genomes.
- Identify GC-rich regulatory regions.

GC content is one of the first sequence-level metrics reported in many genomics studies.

# **Manual Calculation of GC Content**

The GC content of a DNA sequence is calculated using the following formula:

**GC Content (%) = [(Number of G + Number of C) ÷ Total Number of Nucleotides] × 100**

**Example**

DNA Sequence:

ATGCGC

Number of G = 2

Number of C = 2

Total Length = 6

GC Content = (2 + 2) / 6 × 100 = 66.67%

Although this calculation can be performed manually for short sequences, Python enables rapid analysis of thousands of DNA sequences.

````python
#creating a dna sequence
# TP53-inspired DNA sequence (educational example)

dna = "ATGGCTAGCTTGACCGATGCTAGCGTACCGGATCGTACGATCGTAGCTAG"

print("DNA Sequence:")

print(dna)
````

````python
# Using the DNA sequence created earlier

print("DNA Sequence:")
print(dna)

G = dna.count("G")
C = dna.count("C")

print("\nNumber of G:", G)
print("Number of C:", C)
````

# **Understanding the Code**

The `count()` function counts the occurrences of Guanine (G) and Cytosine (C) within the DNA sequence.

These values are then used to calculate the GC content percentage.

Separating the counts also helps verify intermediate calculations before computing the final percentage.

````python
#Calculate GC Content
length = len(dna)

gc_content = ((G + C) / length) * 100

print("GC Content:", round(gc_content,2), "%")
````

# **Understanding the Calculation**

The total number of G and C nucleotides is divided by the total sequence length.

The result is multiplied by 100 to express GC content as a percentage.

The `round()` function limits the output to two decimal places, making the result easier to read.

# **Biological Interpretation**

GC content provides insight into DNA composition.

General guidelines:

- **Below 40%** → AT-rich sequence
- **40–60%** → Balanced sequence
- **Above 60%** → GC-rich sequence

GC-rich DNA tends to be more thermally stable because G–C base pairs contain three hydrogen bonds, whereas A–T pairs contain only two.

# **GC Content and DNA Stability**

DNA stability is influenced by base pairing.

| Base Pair | Hydrogen Bonds |
|-----------|----------------|
| A – T | 2 |
| G – C | 3 |

Because G–C pairs contain an additional hydrogen bond, GC-rich DNA requires more energy to separate during biological processes such as DNA replication, transcription, and PCR.

Consequently, GC-rich DNA generally exhibits a higher melting temperature than AT-rich DNA.

````python
#Compare Multiple DNA Sequences
sequences = {

    "Sequence 1":"ATGGCTAGCTTGACCGATGCTAGCGTACCGG",

    "Sequence 2":"ATATATATATATATATATATATATAT",

    "Sequence 3":"GCGCGCGCGCGCGCGCGCGCGCGCGC"

}

for name, seq in sequences.items():

    gc = ((seq.count("G") + seq.count("C")) / len(seq)) * 100

    print(name)

    print("GC Content:", round(gc,2), "%")

    print()
````

# **Interpretation**

This comparison demonstrates how GC content varies among different DNA sequences.

Sequence 2 is expected to have a very low GC content because it contains mostly A and T.

Sequence 3 is expected to have an extremely high GC content because it contains almost exclusively G and C.

Such comparisons are routinely performed in comparative genomics.

````python
#Visualize GC Content
import matplotlib.pyplot as plt

names = []

gc_values = []

for name, seq in sequences.items():

    names.append(name)

    gc_values.append(
        ((seq.count("G")+seq.count("C"))/len(seq))*100
    )

plt.figure(figsize=(8,5))

plt.bar(names, gc_values)

plt.title("Comparison of GC Content")

plt.xlabel("DNA Sequence")

plt.ylabel("GC Content (%)")

plt.show()
````

# **Understanding the Figure**

The bar chart compares the GC content of multiple DNA sequences.

Higher bars indicate greater GC content.

Visual comparisons help researchers rapidly identify GC-rich and AT-rich regions within genomic datasets.

# **Base Pairing Rules**

**Complementary Base Pairing Rules**

DNA follows Chargaff's Rule:

| DNA Base | Complement |
|-----------|------------|
| A | T |
| T | A |
| G | C |
| C | G |

These rules allow cells to accurately copy genetic information during DNA replication and transcription.

````python
#Original DNA

dna = "ATGGCTAGCTTGACCGATGCTAGCGTACCGGATCGTACGATCGTAGCTAG"

print("Original DNA Sequence:")
print(dna)
````

````python
complement = dna.translate(
    str.maketrans("ATGC", "TACG")
)

print("Complementary DNA:")
print(complement)
````

# **Understanding the Code**

The `translate()` function replaces each nucleotide according to the complementary base-pairing rules.

A → T

T → A

G → C

C → G

Python performs this transformation automatically for every nucleotide in the sequence.

# **Biological Interpretation**

The complementary strand represents the second strand of the DNA double helix.

Inside living cells, DNA replication depends on complementary base pairing to accurately duplicate genetic information.

Computational generation of complementary sequences is frequently used during sequence analysis and genome annotation.

````python
#Reverse Complement
reverse_complement = complement[::-1]

print("Reverse Complement:")
print(reverse_complement)
````

The computational comparison demonstrated here represents the first step of many variant-calling pipelines.

````python
#Normal vs. Mutated DNA
normal_dna = "ATGGCTAGCTTGACCGATGCTAGCGTACCGG"

mutated_dna = "ATGGCTAGCTTGATCGATGCTAGCGTACCGG"

print("Normal DNA:")
print(normal_dna)

print()

print("Mutated DNA:")
print(mutated_dna)
````

# **Explanation**

### **Understanding the Sequences**

The two DNA sequences are almost identical.

However, one nucleotide has been changed.

Our goal is to identify:

- whether a mutation exists,
- where it occurs,
- and what nucleotide changed.

````python
#Compare Position by Position
print("Mutation Analysis")

for i in range(len(normal_dna)):

    if normal_dna[i] != mutated_dna[i]:

        print(f"Mutation detected at Position {i}")

        print(f"Normal : {normal_dna[i]}")

        print(f"Mutated: {mutated_dna[i]}")
````

# **Explanation**

### **How the Comparison Works ?**

The program examines both sequences one nucleotide at a time.

If the nucleotides differ at the same position, Python reports:

- mutation position,
- original nucleotide,
- mutated nucleotide.

This simple comparison demonstrates the principle behind sequence alignment and variant detection.

# **Biological Interpretation**

A single nucleotide substitution may alter:

- protein sequence,
- protein function,
- gene regulation,
- disease susceptibility.

Although many substitutions are harmless, others may contribute to diseases including cancer.

Determining the biological impact requires additional analyses beyond sequence comparison.

````python
#Mutation Summary Table
import pandas as pd

mutation_summary = []

for i in range(len(normal_dna)):

    if normal_dna[i] != mutated_dna[i]:

        mutation_summary.append({

            "Position": i,

            "Normal": normal_dna[i],

            "Mutated": mutated_dna[i],

            "Mutation Type": "Substitution"

        })

mutation_df = pd.DataFrame(mutation_summary)

mutation_df
````

# **Deletion Mutation**

Deletion mutations remove nucleotides from a DNA sequence.

Depending on their location, deletions may shorten proteins or produce frameshift mutations that alter downstream amino acid sequences.

````python
#Compare Multiple Patient Samples
patients = {

    "Patient_1":"ATGGCTAGCTTGACCGATGCTAGCGTACCGG",

    "Patient_2":"ATGGCTAGCTTGATCGATGCTAGCGTACCGG",

    "Patient_3":"ATGGCTAGCTTGACCGATGCTAGCGTATCGG"

}

reference = patients["Patient_1"]

for patient, sequence in patients.items():

    print("\n", patient)

    mutation_found = False

    for i in range(len(reference)):

        if reference[i] != sequence[i]:

            mutation_found = True

            print(
                f"Mutation at position {i}: "
                f"{reference[i]} → {sequence[i]}"
            )

    if not mutation_found:

        print("No mutations detected.")
````

# **Biological Interpretation**

By comparing multiple patient samples with a reference sequence, researchers can identify which individuals harbor sequence variants.

This approach is commonly used in:

- cancer genomics,
- infectious disease surveillance,
- inherited disease screening,
- evolutionary biology.

````python
#Mutation Summary Across Patients
summary = []

for patient, sequence in patients.items():

    mutations = 0

    for i in range(len(reference)):

        if reference[i] != sequence[i]:

            mutations += 1

    summary.append({

        "Patient": patient,

        "Number_of_Mutations": mutations

    })

summary_df = pd.DataFrame(summary)

summary_df
````

# **Biological Interpretation**

````python
#Nucleotide Composition & GC Content
results = []

for patient, sequence in patients.items():

    A = sequence.count("A")

    T = sequence.count("T")

    G = sequence.count("G")

    C = sequence.count("C")

    GC = round(((G+C)/len(sequence))*100,2)

    results.append({

        "Patient":patient,

        "Length":len(sequence),

        "A":A,

        "T":T,

        "G":G,

        "C":C,

        "GC_Content":GC

    })

summary = pd.DataFrame(results)

summary
````
