```python
#Simple Function
def welcome():
    print("Welcome to Computational Biology")

welcome()
```

```python
#Function with Parameter
def gene_name(name):
    print("Gene:", name)

gene_name("TP53")
```

```python
#Return
def dna_length(sequence):

    return len(sequence)

length = dna_length("ATGCGATACCG")

print(length)
```

```python
#GC Content Function
def gc_content(sequence):

    g = sequence.count("G")
    c = sequence.count("C")

    gc = ((g+c)/len(sequence))*100

    return gc

dna = "ATGCGATACCG"

print(gc_content(dna))
```

```python
#Function for Reverse DNA
def reverse_sequence(sequence):

    return sequence[::-1]

dna = "ATGCGAT"

print(reverse_sequence(dna))
```

```python
#Mini Project
def nucleotide_count(sequence):

    print("A =", sequence.count("A"))
    print("T =", sequence.count("T"))
    print("G =", sequence.count("G"))
    print("C =", sequence.count("C"))

dna = "ATGCGATACCG"

nucleotide_count(dna)
```
