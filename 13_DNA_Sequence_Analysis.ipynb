````python
with open("Cancer_Genes.fasta", "r") as file:

    lines = file.readlines()

print("Total Lines in FASTA File:", len(lines))
````

# **Understanding the Code**

The `readlines()` function reads every line of the FASTA file and stores them as a Python list.

Each element in the list corresponds to one line from the FASTA file.


# **Display the File Line by Line**

````python
for line in lines:

    print(line.strip())
````

# **Understanding the Output**

Notice that every FASTA entry consists of:

Header

```
>TP53_Human
```

Sequence

```
ATGGCTAGCTTGACCGATGCTAGCGTACCGG
```

The header always begins with the `>` symbol, making it easy for Python to distinguish sequence names from nucleotide sequences.

````python
sequences = {}

current_header = ""

for line in lines:

    line = line.strip()

    if line.startswith(">"):

        current_header = line[1:]

        sequences[current_header] = ""

    else:

        sequences[current_header] += line

print(sequences)
````

# **Explanation**

**How FASTA Parsing Works**

This code converts the FASTA file into a Python dictionary.

Dictionary Structure

```
Header → DNA Sequence
```

Example

```
TP53_Human

↓

ATGGCTAGCTTGACCGATGCTAGCGTACCGG
```

Using dictionaries makes it easy to retrieve sequences by their identifiers.

````python
#Create a Summary Table
import pandas as pd

summary = []

for header, sequence in sequences.items():

    summary.append({

        "Gene": header,

        "Length": len(sequence),

        "A": sequence.count("A"),

        "T": sequence.count("T"),

        "G": sequence.count("G"),

        "C": sequence.count("C")

    })

summary_df = pd.DataFrame(summary)

summary_df
````

# **Section 13D – Comparative GC Content Analysis of FASTA Sequences**
After extracting DNA sequences from a FASTA file, one of the first quantitative analyses performed in computational genomics is the calculation of **GC content**.

GC content represents the percentage of guanine (G) and cytosine (C) nucleotides present in a DNA sequence.

Because guanine and cytosine form **three hydrogen bonds**, whereas adenine and thymine form only **two hydrogen bonds**, GC-rich DNA is generally more stable.

In this section, we will calculate the GC content of every sequence contained in our FASTA file, compare sequences quantitatively, visualize the results, and interpret their biological significance.

# **Learning Objectives**

# **Why GC Content Matters?**

GC content is one of the most widely reported characteristics of DNA sequences.

Researchers calculate GC content to:

- Compare genomes.
- Assess sequence composition.
- Evaluate DNA stability.
- Design PCR primers.
- Characterize chromosomes.
- Study microbial evolution.
- Analyze transcriptomes.
- Identify unusual genomic regions.

Almost every genomics study includes GC content as a basic descriptive statistic.

````python
#Calculate GC Content for Every Sequence

gc_contents = []

for header, sequence in sequences.items():

    G = sequence.count("G")

    C = sequence.count("C")

    gc = ((G + C) / len(sequence)) * 100

    gc_contents.append(round(gc, 2))

gc_contents
````

# **Explanation**

**Understanding the Code**

For every DNA sequence, Python:

1. Counts the number of **G** nucleotides.
2. Counts the number of **C** nucleotides.
3. Divides their sum by the sequence length.
4. Multiplies by 100 to calculate GC percentage.

The result is a list containing the GC content of every sequence in the FASTA file.

# **Add GC Content to the Summary Table**

````python
summary_df["GC_Content (%)"] = gc_contents

summary_df
````

# **GC Content and DNA Stability**

**Relationship Between GC Content and DNA Stability**

DNA stability depends on hydrogen bonding.

| Base Pair | Hydrogen Bonds |
|-----------|----------------|
| A–T | 2 |
| G–C | 3 |

Because G–C pairs contain an additional hydrogen bond, GC-rich DNA generally:

- Requires higher temperatures to separate.
- Has higher melting temperatures.
- Exhibits greater structural stability.

These characteristics influence PCR amplification, sequencing efficiency, and molecular diagnostics.

````python
#Search for Start Codons (ATG)

print("Searching for Start Codons (ATG)\n")

for gene, sequence in sequences.items():

    if "ATG" in sequence:

        print(f"{gene}: Start codon detected")

    else:

        print(f"{gene}: Start codon NOT detected")
````

### **Understanding the Code**

The `in` operator checks whether the motif **ATG** exists within each DNA sequence.

ATG is the most common **start codon**, signaling where protein translation begins.

Detecting start codons is an important step in identifying potential protein-coding regions.

# **Biological Interpretation**

Genes containing an ATG start codon may represent protein-coding sequences.

During translation, the ribosome recognizes this codon and initiates protein synthesis.

Although not every ATG functions as a true start site, its presence provides an initial clue during genome annotation.

````python
#Search for Additional Motifs

motifs = ["ATG", "TATA", "CGG"]

for motif in motifs:

    print(f"\nSearching for motif: {motif}")

    for gene, sequence in sequences.items():

        if motif in sequence:

            print(f"{gene}: Present")

        else:

            print(f"{gene}: Not Present")
````

````python
#Load the FASTA Dataset

# Read the FASTA file

with open("Cancer_Genes.fasta", "r") as file:
    fasta_lines = file.readlines()

print("Number of lines in FASTA file:", len(fasta_lines))
````

# **Explanation**

The `readlines()` function loads every line of the FASTA file into memory.

Each line corresponds either to:

- a sequence header beginning with `>`
- or a DNA sequence.

This is the first step in converting raw biological sequence files into structured datasets suitable for computational analysis.

````python
#Parse the FASTA File

sequences = {}

current_gene = ""

for line in fasta_lines:

    line = line.strip()

    if line.startswith(">"):

        current_gene = line[1:]

        sequences[current_gene] = ""

    else:

        sequences[current_gene] += line

print("Number of sequences:", len(sequences))
````

# **Explanation**

**Understanding FASTA Parsing**

FASTA parsing separates sequence identifiers (headers) from the nucleotide sequences.

After parsing, the data are stored in a Python dictionary:

Gene Name → DNA Sequence

This structure enables efficient retrieval and analysis of each sequence.

````python
#Generate Sequence Statistics

import pandas as pd

project_summary = []

for gene, sequence in sequences.items():

    length = len(sequence)

    A = sequence.count("A")
    T = sequence.count("T")
    G = sequence.count("G")
    C = sequence.count("C")

    gc = round(((G + C) / length) * 100, 2)

    atg = sequence.count("ATG")

    project_summary.append({

        "Gene": gene,

        "Length": length,

        "A": A,

        "T": T,

        "G": G,

        "C": C,

        "GC_Content (%)": gc,

        "ATG_Count": atg

    })

project_df = pd.DataFrame(project_summary)

project_df
````

# **Explanation**

### **Project Summary Table**

The resulting DataFrame combines several sequence characteristics into a single table.

For every gene we now know:

- sequence length
- nucleotide composition
- GC percentage
- number of ATG motifs

This resembles the summary tables commonly generated during exploratory genomics analyses.
