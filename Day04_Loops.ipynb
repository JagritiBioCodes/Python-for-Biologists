```python
print("Welcome to Day 4")
print("Learning Python Loops")
```

```python
#For loop
genes = ["TP53","BRCA1","EGFR","KRAS","MYC"]

for gene in genes:
    print(gene)
```

```python
#Loop with Numbering
genes = ["TP53","BRCA1","EGFR","KRAS","MYC"]

for i, gene in enumerate(genes, start=1):
    print(i, gene)
```

```python
#Range()
for sample in range(1,6):
    print("Sample", sample)
```

```python
#Loop through DNA
dna = "ATGCGATACCGT"

for nucleotide in dna:
    print(nucleotide)
```

```python
#Count G
dna = "ATGCGATACCGT"

count = 0

for base in dna:
    if base == "G":
        count += 1

print("Total G =", count)
```

```python
#While Loop
reads = 1

while reads <= 5:
    print("Sequencing Run", reads)
    reads += 1
```

```python
#Break
genes = ["TP53","BRCA1","EGFR","KRAS"]

for gene in genes:

    if gene == "EGFR":
        break

    print(gene)
```

```python
#Continue
genes = ["TP53","BRCA1","EGFR","KRAS"]

for gene in genes:

    if gene == "EGFR":
        continue

    print(gene)
```

```python
#Mini Project
dna = "ATGCGATACCGT"

for base in dna:
    print(base)
```
