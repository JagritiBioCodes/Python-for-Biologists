# **What is File Handling?**
File Handling refers to reading, writing, updating, and managing files stored on a computer.

Instead of typing data manually every time, researchers store biological information in files.

Python allows us to access these files, analyze their contents, and generate new output files automatically.

This is a fundamental step in every bioinformatics pipeline.

```python
#Example 1 Writing a DNA Sequence
dna = "ATGCGTAGCTAGCTAGCTAG"

with open("dna_sequence.txt", "w") as file:
    file.write(dna)

print("DNA sequence saved successfully.")
```

# **Explanation**
The file was opened in **write mode ("w")**.

Python created a new text file named:

dna_sequence.txt

The DNA sequence was written into this file.

If the file already existed, its previous contents would be replaced.

# **Biological Interpretation**
DNA sequences obtained from sequencing experiments are often stored in files.

Although real sequencing data usually uses FASTA or FASTQ formats, understanding how to write a simple text file is the first step toward handling biological sequence data.

```python
#Example 2 Reading a DNA Sequence
with open("dna_sequence.txt", "r") as file:
    sequence = file.read()

print(sequence)
```

# **Explanation**
The file was opened in **read mode ("r")**.

The complete contents of the file were read into the variable:

sequence

The sequence was then printed.

# **Biological Interpretation**
Researchers frequently read DNA sequences from files before calculating:

• Sequence length

• GC content

• Nucleotide frequency

• Protein translation

• Mutation detection

```python
#Example 3 Appending Additional DNA
with open("dna_sequence.txt", "a") as file:
    file.write("\nTTAGCGATCG")

print("New DNA sequence appended.")
```

```python
#Read Again
with open("dna_sequence.txt", "r") as file:
    print(file.read())
```

# **Explanation**
Append mode ("a") adds new information to the end of an existing file.

Unlike write mode, append mode does not delete previous data.

This is useful when sequencing experiments produce new samples over time.

```python
#Biological Example -Saving Gene Names
genes = [

    "TP53",

    "BRCA1",

    "EGFR",

    "KRAS",

    "MYC"

]

with open("genes.txt", "w") as file:

    for gene in genes:

        file.write(gene + "\n")

print("Gene list saved successfully.")
```

```python
#Reading Gene List
with open("genes.txt", "r") as file:

    print(file.read())
```

# **Biological Interpretation**
Gene lists are generated in many biological studies, including:

• RNA-Seq

• Differential Gene Expression

• GWAS

• Cancer Genomics

Saving gene names into files allows researchers to perform downstream analyses using other bioinformatics tools.

```python
#Reading Line by Line
with open("genes.txt", "r") as file:

    for gene in file:

        print(gene.strip())
```

# **Explanation**
The strip() function removes extra newline characters.

Reading files line by line is much more memory efficient for large datasets.

Many FASTA and FASTQ parsers use this approach.

# **Mini Exercise**
A researcher identified five biomarkers in a breast cancer study.

Biomarkers:

HER2

TP53

PIK3CA

AKT1

PTEN

Tasks:

1. Store the biomarkers in a Python list.

2. Save them into a file named biomarkers.txt.

3. Read the file.

4. Print each biomarker line by line.

```python
#Solution
biomarkers = [

    "HER2",

    "TP53",

    "PIK3CA",

    "AKT1",

    "PTEN"

]

with open("biomarkers.txt", "w") as file:

    for marker in biomarkers:

        file.write(marker + "\n")

print("Biomarkers saved.")

print("\nReading File:\n")

with open("biomarkers.txt", "r") as file:

    for marker in file:

        print(marker.strip())
```

* Forgetting that write mode overwrites existing content
