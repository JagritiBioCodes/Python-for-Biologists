````python
#Create Simulated FASTA Dataset
fasta_sequences = {

    "TP53":
    "ATGGAGGAGCCGCAGTCAGATCCTAGCG",

    "EGFR":
    "ATGCGACCCTCCGGGACGGCCTCTTCGG",

    "KRAS":
    "ATGACTGAATATAAACTTGTGGTAGTTG",

    "BRCA1":
    "ATGTCTCTGTGTGACAAAGCCAGGGAGC",

    "MYC":
    "ATGCCCCTCAACGTTAGCTTCACCAACA"

}

fasta_sequences
````

# **Explanation**

**Simulated FASTA Dataset**

Five LUAD-associated genes are represented as DNA sequences.

These sequences will be analyzed computationally throughout this section.

````python
#Sequence Length
lengths = {}

for gene, sequence in fasta_sequences.items():

    lengths[gene] = len(sequence)

lengths
````

# **Biological Interpretation**

Sequence length is one of the first quality-control metrics examined during sequence analysis.

Longer sequences generally contain more biological information and may encode larger protein domains.

````python
#Nucleotide Composition
composition = {}

for gene, sequence in fasta_sequences.items():

    composition[gene] = {

        "A": sequence.count("A"),

        "T": sequence.count("T"),

        "G": sequence.count("G"),

        "C": sequence.count("C")

    }

composition
````

# **Explanation**

**Nucleotide Composition**

Counting nucleotides provides insight into DNA composition and forms the basis for GC content analysis.

````python
#GC Content
gc_content = {}

for gene, sequence in fasta_sequences.items():

    gc = (
        sequence.count("G") +
        sequence.count("C")
    ) / len(sequence) * 100

    gc_content[gene] = round(gc,2)

gc_content
````

# **Biological Interpretation**

GC-rich DNA regions generally exhibit greater thermal stability due to the presence of three hydrogen bonds between guanine and cytosine.

GC content is widely used in comparative genomics, PCR primer design, and genome annotation.

````python
#Motif Search
motifs = ["ATG","TATA","CGG"]

motif_results = {}

for gene, sequence in fasta_sequences.items():

    motif_results[gene] = {

        motif: sequence.count(motif)

        for motif in motifs

    }

motif_results
````

# **Biological Interpretation**

Motifs are short conserved DNA patterns associated with biological function.

Examples include:

- **ATG** – Translation start codon.
- **TATA** – Promoter-associated motif.
- **CGG** – GC-rich regulatory sequence.

````python
#Reverse Complement
complement = {

    "A":"T",

    "T":"A",

    "G":"C",

    "C":"G"

}

reverse_complements = {}

for gene, sequence in fasta_sequences.items():

    reverse = "".join(

        complement[base]

        for base in sequence[::-1]

    )

    reverse_complements[gene] = reverse

reverse_complements
````

````python
#RNA Transcription
rna_sequences = {}

for gene, sequence in fasta_sequences.items():

    rna_sequences[gene] = sequence.replace("T","U")

rna_sequences
````

# **Explanation**

**RNA Transcription**

DNA transcription converts thymine (T) into uracil (U), producing an RNA sequence.

This process represents the first step of gene expression in the central dogma of molecular biology.

````python
#Mutation Detection
reference = fasta_sequences["TP53"]

mutation_summary = {}

for gene, sequence in fasta_sequences.items():

    differences = 0

    for a,b in zip(reference,sequence):

        if a != b:

            differences += 1

    mutation_summary[gene] = differences

mutation_summary
````

# **Explanation**

**Mutation Detection**

Each sequence is compared with the TP53 reference sequence.

The number of differing nucleotide positions provides a simple measure of sequence divergence.

More sophisticated analyses would use sequence alignment algorithms such as Needleman–Wunsch or Smith–Waterman.

````python
#Integrated Summary Table
summary = pd.DataFrame({

    "Gene": list(fasta_sequences.keys()),

    "Length":[lengths[g] for g in fasta_sequences],

    "GC_Content":[gc_content[g] for g in fasta_sequences],

    "ATG_Motif":[motif_results[g]["ATG"] for g in fasta_sequences],

    "TATA_Motif":[motif_results[g]["TATA"] for g in fasta_sequences],

    "CGG_Motif":[motif_results[g]["CGG"] for g in fasta_sequences],

    "Mutations_vs_TP53":[mutation_summary[g] for g in fasta_sequences]

})

summary
````

````python
#Visualization
plt.figure(figsize=(8,5))

plt.bar(

    summary["Gene"],

    summary["GC_Content"]

)

plt.title("GC Content Across LUAD Genes")

plt.xlabel("Gene")

plt.ylabel("GC Content (%)")

plt.show()
````

# **Biological Interpretation**

Comparing GC content across genes highlights differences in sequence composition that may influence DNA stability, amplification efficiency, and gene-specific characteristics.

Although this dataset is simulated, the workflow reflects analyses routinely performed during sequence characterization and comparative genomics studies.

````python
#Display Final Summary
summary.sort_values(by="GC_Content",ascending=False)
````
