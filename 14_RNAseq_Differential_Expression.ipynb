DGE analysis helps researchers identify genes that may contribute to disease progression, serve as diagnostic biomarkers, or represent potential therapeutic targets.

# **Understanding Log₂ Fold Change (Log₂FC)**

Log₂ Fold Change measures the magnitude of expression change between two biological groups.

Interpretation:

**Log₂FC > 0**

→ Gene expression increases

→ Upregulated



**Log₂FC < 0**

→ Gene expression decreases

→ Downregulated


Example

| Log₂FC | Interpretation |
|---------|---------------|
| +3 | Strongly Upregulated |
| +1 | Mildly Upregulated |
| 0 | No Change |
| -1 | Mildly Downregulated |
| -3 | Strongly Downregulated |

Researchers often use thresholds such as:

Log₂FC ≥ 1

Log₂FC ≤ -1

to identify biologically meaningful expression changes.

# **Understanding P-value**

The P-value estimates the probability that the observed expression difference occurred by chance.

Smaller P-values indicate stronger statistical evidence.

**Typical interpretation**:

**P-value < 0.05**

**↓**

Statistically Significant

However, RNA-Seq experiments test thousands of genes simultaneously.

Therefore, P-values alone are insufficient because multiple hypothesis testing increases the chance of false positives.

# **Understanding Adjusted P-value (False Discovery Rate)**

The Adjusted P-value (False Discovery Rate, FDR) corrects for multiple statistical comparisons.

Most transcriptomic studies use:

**Adjusted P-value < 0.05**

to define statistically significant genes.

FDR greatly reduces false-positive discoveries and improves the reliability of differential expression analysis.

````python
#Load Dataset
import pandas as pd

luad_df = pd.read_csv("LUAD_Gene_Expression_Dataset.csv")

luad_df.head()
````

````python
#Identify Significantly Upregulated Genes
upregulated = luad_df[
    (luad_df["Log2FC"] >= 1) &
    (luad_df["Adjusted_P_Value"] < 0.05)
]

upregulated.head()
````

# **Explanation**

**Understanding the Filter**

This code identifies genes satisfying both criteria:

* Log₂ Fold Change ≥ 1, or

* Adjusted P-value < 0.05

Only genes passing both conditions are considered significantly upregulated.

# **Biological Interpretation**
Genes identified in this table exhibit significantly increased expression in the simulated LUAD dataset.

Such genes may function as:

- Oncogenes

# **Section 14F.2 – Publication-Quality Volcano Plot**
One of the most widely used visualizations in transcriptomics is the **Volcano Plot**.

A volcano plot simultaneously displays:

- Magnitude of gene expression change (Log₂ Fold Change)
- Statistical significance (-log₁₀ Adjusted P-value)

This allows researchers to rapidly identify genes that show both:

- Large expression differences
- Strong statistical significance

Genes appearing in the upper left and upper right regions are generally the most biologically interesting because they represent significantly downregulated and upregulated genes.

# **Biological Importance**
**Why Volcano Plots Matter?**

RNA-Seq experiments often evaluate thousands of genes simultaneously.

Reading tables alone makes it difficult to identify important genes.

Volcano plots summarize differential expression results into a single figure by displaying:

• Fold Change

• Statistical Significance

Researchers routinely include volcano plots in publications because they immediately highlight candidate biomarkers and therapeutic targets.

````python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
#Load Dataset
luad_df = pd.read_csv("LUAD_Gene_Expression_Dataset.csv")

luad_df.head()
````

````python
#Prepare the Dataset
# Create Absolute Log2 Fold Change
luad_df["Absolute_Log2FC"] = luad_df["Log2FC"].abs()

# Calculate -log10(FDR)
luad_df["MinusLog10FDR"] = -np.log10(
    luad_df["Adjusted_P_Value"]
)

luad_df.head()
````

# **Understanding the Code**

Two new variables are created:

**Absolute_Log2FC**

Measures the magnitude of expression change regardless of direction.

**MinusLog10FDR**

Converts very small Adjusted P-values into larger positive values.

Smaller Adjusted P-values become higher points on the graph, making significant genes easier to identify.

````python
#Classify Genes
conditions = [

    (luad_df["Log2FC"] >= 1) &
    (luad_df["Adjusted_P_Value"] < 0.05),

    (luad_df["Log2FC"] <= -1) &
    (luad_df["Adjusted_P_Value"] < 0.05)

]

choices = [

    "Upregulated",

    "Downregulated"

]

luad_df["Category"] = np.select(
    conditions,
    choices,
    default="Not Significant"
)

luad_df.head()
````

# **Explanation**

**Gene Classification**

Each gene is classified into one of three categories:

🟥 Upregulated

🟦 Downregulated

⚪ Not Significant

These categories will be represented using different colors in the volcano plot.

````python
#Create the Volcano Plot
plt.figure(figsize=(10,8))

# Non-significant genes
plt.scatter(
    luad_df[luad_df["Category"]=="Not Significant"]["Log2FC"],
    luad_df[luad_df["Category"]=="Not Significant"]["MinusLog10FDR"],
    color="lightgray",
    alpha=0.6,
    label="Not Significant"
)

# Upregulated genes
plt.scatter(
    luad_df[luad_df["Category"]=="Upregulated"]["Log2FC"],
    luad_df[luad_df["Category"]=="Upregulated"]["MinusLog10FDR"],
    color="red",
    alpha=0.8,
    label="Upregulated"
)

# Downregulated genes
plt.scatter(
    luad_df[luad_df["Category"]=="Downregulated"]["Log2FC"],
    luad_df[luad_df["Category"]=="Downregulated"]["MinusLog10FDR"],
    color="blue",
    alpha=0.8,
    label="Downregulated"
)

# Threshold lines
plt.axvline(x=1,
            linestyle="--",
            color="black")

plt.axvline(x=-1,
            linestyle="--",
            color="black")

plt.axhline(
    y=-np.log10(0.05),
    linestyle="--",
    color="black"
)

plt.xlabel("Log₂ Fold Change")

plt.ylabel("-log₁₀ Adjusted P-value")

plt.title("Volcano Plot of Simulated LUAD RNA-Seq Dataset")

plt.legend()

plt.show()
````

# **Explanation**

**Understanding the Volcano Plot**

**X-axis**

Represents Log₂ Fold Change.

Genes on the right are upregulated.

Genes on the left are downregulated.



**Y-axis**

Represents statistical significance.

Higher points indicate smaller Adjusted P-values and therefore stronger evidence for differential expression.



**Dashed Vertical Lines**

Represent the Log₂ Fold Change thresholds:

Log₂FC = +1

Log₂FC = -1


**Dashed Horizontal Line**

Represents the statistical significance threshold:

Adjusted P-value = 0.05

````python
#Annotate the Top 10 Biomarkers
top10 = luad_df.nsmallest(
    10,
    "Adjusted_P_Value"
)

plt.figure(figsize=(10,8))

colors = {
    "Upregulated":"red",
    "Downregulated":"blue",
    "Not Significant":"gray"
}

for category in colors:

    subset = luad_df[
        luad_df["Category"]==category
    ]

    plt.scatter(
        subset["Log2FC"],
        subset["MinusLog10FDR"],
        color=colors[category],
        alpha=0.7,
        label=category
    )

# Label Top 10 Genes
for _, row in top10.iterrows():

    plt.text(
        row["Log2FC"],
        row["MinusLog10FDR"],
        row["Gene"],
        fontsize=8
    )

plt.axvline(1,
            linestyle="--",
            color="black")

plt.axvline(-1,
            linestyle="--",
            color="black")

plt.axhline(
    -np.log10(0.05),
    linestyle="--",
    color="black"
)

plt.xlabel("Log₂ Fold Change")

plt.ylabel("-log₁₀ Adjusted P-value")

plt.title("Publication-Style Volcano Plot")

plt.legend()

plt.show()
````

# **Biological Importance**

**Why Multi-Sample Heatmaps Matter**?

Unlike simple heatmaps, transcriptomic heatmaps compare expression across many patients simultaneously.

Researchers use these figures to identify:

- Candidate biomarkers
- Disease-associated expression signatures
- Patient-specific variation
- Co-regulated genes
- Molecular subtypes

This type of visualization is routinely presented in TCGA and GEO publications.

````python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
````

````python
#Load Dataset
luad_df = pd.read_csv("LUAD_Gene_Expression_Dataset.csv")

luad_df.head()
````

````python
#Select Top Differentially Expressed Genes
top20 = luad_df.sort_values(
    by="Adjusted_P_Value"
).head(20)

top20
````

# **Explanation**

The twenty genes with the smallest Adjusted P-values are selected.

These represent the strongest candidate biomarkers for visualization.

````python
#Simulate RNA-Seq Expression Matrix
expression_matrix = pd.DataFrame(index=top20["Gene"])

# Normal Samples
for i in range(1,6):

    expression_matrix[f"Normal_{i}"] = np.round(
        np.random.normal(
            top20["Expression"],
            1.5
        ),
        2
    )

# Tumor Samples
for i in range(1,6):

    expression_matrix[f"Tumor_{i}"] = np.round(
        np.random.normal(
            top20["Expression"] + top20["Log2FC"]*2,
            1.5
        ),
        2
    )

expression_matrix.head()
````

# **Explanation**

**Simulating Patient Samples**

Each column represents one patient.

Five columns represent normal lung tissue.

Five columns represent LUAD tumor tissue.

Tumor samples are simulated by increasing or decreasing expression according to the Log₂ Fold Change.

This creates a realistic transcriptomic expression matrix.

````python
#Generate Publication-Style Heatmap
plt.figure(figsize=(10,8))

plt.imshow(
    expression_matrix,
    aspect="auto",
    interpolation="nearest"
)

plt.colorbar(label="Normalized Expression")

plt.xticks(
    range(len(expression_matrix.columns)),
    expression_matrix.columns,
    rotation=45
)

plt.yticks(
    range(len(expression_matrix.index)),
    expression_matrix.index
)

plt.title(
    "Expression Heatmap of Top LUAD Biomarkers"
)

plt.xlabel("Patient Samples")

plt.ylabel("Genes")

plt.tight_layout()

plt.show()
````

# **Explanation**

**Understanding the Heatmap**

Rows correspond to genes.

Columns correspond to individual patient samples.

Colors represent normalized gene expression values.

Similar color patterns indicate similar expression profiles.

````python
#Compare Average Expression
comparison = pd.DataFrame({

    "Normal_Average":
        expression_matrix.iloc[:,0:5].mean(axis=1),

    "Tumor_Average":
        expression_matrix.iloc[:,5:10].mean(axis=1)

})

comparison.head()
````

````python
#Visualize Average Expression
comparison.plot(
    kind="bar",
    figsize=(12,5)
)

plt.ylabel("Average Expression")

plt.title(
    "Average Gene Expression in Normal vs Tumor Samples"
)

plt.tight_layout()

plt.show()
````

Genes with lower tumor expression may represent:
